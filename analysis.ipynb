```python
# parse Date
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# if you have a clock column like "TimeUnder" as "MM:SS" or "M:SS"
def mmss_to_seconds(s):
    try:
        m, sec = str(s).split(':')
        return int(m)*60 + int(sec)
    except Exception:
        return pd.NA

if 'TimeUnder' in df.columns:
    df['TimeUnderSecs'] = df['TimeUnder'].apply(mmss_to_seconds).astype('Int64')
```

```python
df.head(10)
```

```python
num_cols = df.columns[df.dtypes.isin(['float64','int64'])].tolist()

# also coerce columns that are strings but numeric
possible_numeric = ['time','TimeSecs','PlayTimeDiff']  # extend as needed
for c in possible_numeric:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')
```

```python
df.head(10)
```

```python
# (you already run drop_duplicates earlier) re-check
print("duplicates:", df.duplicated().sum())
# keep first occurrence if duplicates remain
df = df.drop_duplicates(ignore_index=True)
```

```python
missing_frac = df.isna().mean()
drop_cols = missing_frac[missing_frac > 0.8].index.tolist()
print("Dropping columns with >80% missing:", drop_cols)
df = df.drop(columns=drop_cols)
```

```python
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
# or only numeric columns:
# df = df.fillna(df.mean(numeric_only=True))
```

```python
# 1) choose columns you want to forward/backfill and keep only those that exist
cols_to_ffill = [c for c in ['Home_WP_pre','Away_WP_pre','Win_Prob'] if c in df.columns]
if not cols_to_ffill:
    raise KeyError("None of the target columns found in df")

# 2) require ordering columns to exist
if 'GameID' not in df.columns:
    raise KeyError("GameID column missing")
# PlayTimeDiff is recommended for ordering; fall back to index if missing
order_cols = ['GameID', 'PlayTimeDiff'] if 'PlayTimeDiff' in df.columns else ['GameID']

# 3) sort by GameID + ordering and then fill per-group using transform
df = df.sort_values(order_cols).copy()
df[cols_to_ffill] = df.groupby('GameID')[cols_to_ffill].transform(lambda s: s.ffill().bfill())

# 4) (optional) restore original order if you want
# df = df.sort_index()
```

```python
cat_cols = df.select_dtypes(include=['object','category']).columns.tolist()
for c in cat_cols:
    if df[c].isna().any():
        mode = df[c].mode(dropna=True)
        if len(mode):
            df[c] = df[c].fillna(mode.iloc[0])
        else:
            df[c] = df[c].fillna('Unknown')
```

```python
# ensure time is within expected bounds
if 'time' in df.columns:
    df = df[(df['time'].isna()) | (df['time'].between(0, 3600))]

# example: filter plays with non-sensical PlayTimeDiff
if 'PlayTimeDiff' in df.columns:
    df.loc[~df['PlayTimeDiff'].between(-10000,10000), 'PlayTimeDiff'] = pd.NA
```

```python
df[numeric_cols].describe().T
df.isna().sum().sort_values(ascending=False).head(40)
```

```python
# create is_pass target using common column names
possible_playtype_cols = [c for c in ['PlayType','play_type','type','Play'] if c in df.columns]
if possible_playtype_cols:
    col = possible_playtype_cols[0]
    df['is_pass'] = df[col].astype(str).str.lower().str.contains('pass')
    print('is_pass created from', col, '->', df['is_pass'].value_counts(dropna=False).to_dict())
else:
    print('No obvious PlayType column found. Provide the column name or logic to create is_pass.')
```

```python
cols_time_series = [c for c in ['Home_WP_pre','Away_WP_pre','Win_Prob','HomeScore','AwayScore'] if c in df.columns]
order_cols = ['GameID', 'PlayTimeDiff'] if 'PlayTimeDiff' in df.columns else ['GameID']

if cols_time_series:
    df = df.sort_values(order_cols).copy()
    df[cols_time_series] = df.groupby('GameID')[cols_time_series].transform(lambda s: s.ffill().bfill())
    print('Filled time-series columns per GameID:', cols_time_series)
else:
    print('No game-state columns found to forward/backfill.')
```

```python
leakage_patterns = ['post','next','result','after','_post','_next']
leak_cols = [c for c in df.columns if any(p in c.lower() for p in leakage_patterns)]
# keep backup of leakage columns if desired
df_leakage = df[leak_cols].copy() if leak_cols else None
df = df.drop(columns=leak_cols, errors='ignore')
print('Dropped leakage columns:', leak_cols)
```

```python
# coerce likely numeric columns
possible_numeric = ['time','TimeSecs','PlayTimeDiff','yardline','Yards','yards_gained']
for c in possible_numeric:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

numeric_cols = df.select_dtypes(include='number').columns.tolist()
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
print('Numeric fill applied to:', len(numeric_cols), 'columns')
```

```python
# yardline_100
if 'yardline' in df.columns:
    df['yardline_100'] = 100 - df['yardline']

# score diff if scores exist
score_cols = [c for c in ['HomeScore','AwayScore','home_score','away_score'] if c in df.columns]
if len(score_cols) >= 2:
    df['score_diff'] = df[score_cols[0]] - df[score_cols[1]]

# seconds remaining simple approximation (adjust to your clock columns)
if 'TimeSecs' in df.columns and 'qtr' in df.columns:
    df['seconds_remaining_game'] = df['TimeSecs'] + (4 - df['qtr']) * 15 * 60

# recent pass rate per game
df = df.sort_values(['GameID','PlayTimeDiff'] if 'PlayTimeDiff' in df.columns else ['GameID']).copy()
df['last1_is_pass'] = df.groupby('GameID')['is_pass'].shift(1).fillna(0)
df['pass_rate_last5'] = df.groupby('GameID')['is_pass'].transform(lambda s: s.shift(1).rolling(5, min_periods=1).mean()).fillna(0)

print('Added features: yardline_100, score_diff (if available), seconds_remaining_game (if available), last1_is_pass, pass_rate_last5')
```

```python
# Validate targets & class balance
print("=== TARGET VALIDATION ===")
if 'is_pass' in df.columns:
    print("is_pass value counts:")
    print(df['is_pass'].value_counts(dropna=False))
    print("\nis_pass distribution (%):")
    print(df['is_pass'].value_counts(normalize=True) * 100)
    
    # balance by game
    print("\nPass rate by game (sample):")
    game_pass_rate = df.groupby('GameID')['is_pass'].agg(['mean','count']).reset_index()
    print(game_pass_rate.head(10))
else:
    print("WARNING: 'is_pass' not found")

# check for other potential targets
for target in ['Yards','yards_gained','epa','WPA','yacEPA']:
    if target in df.columns:
        print(f"\n{target} stats:")
        print(df[target].describe())
```

```python
# Game-based train/test split (no leakage across games)
from sklearn.model_selection import GroupShuffleSplit

if 'GameID' not in df.columns:
    raise KeyError("GameID missing - cannot split by game")

gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_idx, test_idx = next(gss.split(df, groups=df['GameID']))
df_train = df.iloc[train_idx].reset_index(drop=True)
df_test = df.iloc[test_idx].reset_index(drop=True)

print("=== TRAIN/TEST SPLIT (BY GAME) ===")
print(f"Train: {df_train.shape[0]:,} rows, {df_train['GameID'].nunique()} games")
print(f"Test:  {df_test.shape[0]:,} rows, {df_test['GameID'].nunique()} games")

if 'is_pass' in df.columns:
    print("\nTrain is_pass distribution:")
    print(df_train['is_pass'].value_counts(normalize=True) * 100)
    print("\nTest is_pass distribution:")
    print(df_test['is_pass'].value_counts(normalize=True) * 100)
```

```python
# Baseline Classification: Play-Type (is_pass)
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, accuracy_score
import numpy as np
import sklearn

if 'is_pass' not in df_train.columns:
    print("WARNING: 'is_pass' not found; skipping classification baseline")
else:
    print("=== BASELINE CLASSIFICATION: PLAY TYPE (is_pass) ===\n")
    print(f"sklearn version: {sklearn.__version__}\n")
    
    # Select features: numeric and categorical
    num_features = [c for c in ['down','ydstogo','yardline_100','score_diff','TimeSecs','last1_is_pass','pass_rate_last5'] if c in df_train.columns]
    cat_features = [c for c in ['OffenseTeam','DefenseTeam','SideofField','posteam'] if c in df_train.columns]
    
    print(f"Numeric features ({len(num_features)}): {num_features}")
    print(f"Categorical features ({len(cat_features)}): {cat_features}")
    
    # Preprocessing pipeline - use sparse_output for newer sklearn, sparse for older
    try:
        preprocessor = ColumnTransformer(
            transformers=[
                ('num', StandardScaler(), num_features),
                ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_features)
            ],
            remainder='drop'
        )
    except TypeError:
        preprocessor = ColumnTransformer(
            transformers=[
                ('num', StandardScaler(), num_features),
                ('cat', OneHotEncoder(handle_unknown='ignore', sparse=False), cat_features)
            ],
            remainder='drop'
        )
    
    # Logistic Regression baseline (interpretable)
    clf_lr = Pipeline([
        ('preprocessor', preprocessor),
        ('model', LogisticRegression(max_iter=1000, random_state=42, n_jobs=-1))
    ])
    
    X_train = df_train[num_features + cat_features].copy()
    y_train = df_train['is_pass'].astype(int)
    X_test = df_test[num_features + cat_features].copy()
    y_test = df_test['is_pass'].astype(int)
    
    # Fit & predict
    clf_lr.fit(X_train, y_train)
    y_pred_lr = clf_lr.predict(X_test)
    y_proba_lr = clf_lr.predict_proba(X_test)[:, 1]
    
    print("\n--- LOGISTIC REGRESSION ---")
    print(f"Accuracy: {accuracy_score(y_test, y_pred_lr):.4f}")
    print(f"ROC AUC: {roc_auc_score(y_test, y_proba_lr):.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred_lr, target_names=['Run','Pass']))
    print("\nConfusion Matrix (rows=actual, cols=predicted):")
    print(confusion_matrix(y_test, y_pred_lr))
    
    # Random Forest (better feature importance)
    clf_rf = Pipeline([
        ('preprocessor', preprocessor),
        ('model', RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1, max_depth=10))
    ])
    clf_rf.fit(X_train, y_train)
    y_pred_rf = clf_rf.predict(X_test)
    y_proba_rf = clf_rf.predict_proba(X_test)[:, 1]
    
    print("\n--- RANDOM FOREST ---")
    print(f"Accuracy: {accuracy_score(y_test, y_pred_rf):.4f}")
    print(f"ROC AUC: {roc_auc_score(y_test, y_proba_rf):.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred_rf, target_names=['Run','Pass']))
    
    # Feature importance
    print("\n--- FEATURE IMPORTANCE (RF) ---")
    fi = pd.DataFrame({
        'feature': num_features + cat_features,
        'importance': clf_rf.named_steps['model'].feature_importances_[:len(num_features + cat_features)]
    }).sort_values('importance', ascending=False).head(15)
    display(fi)
```

```python
# Baseline Regression: Outcome (yards or EPA)
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("\n=== BASELINE REGRESSION: OUTCOME PREDICTION ===\n")

# Find regression target
possible_targets = [c for c in ['Yards','yards_gained','epa','WPA','yacEPA'] if c in df_train.columns]
if not possible_targets:
    print("WARNING: No suitable regression target found (Yards, yards_gained, epa, WPA, yacEPA)")
else:
    target = possible_targets[0]
    print(f"Using target: {target}\n")
    
    # Reuse feature set
    num_features = [c for c in ['down','ydstogo','yardline_100','score_diff','TimeSecs','last1_is_pass','pass_rate_last5'] if c in df_train.columns]
    cat_features = [c for c in ['OffenseTeam','DefenseTeam','SideofField','posteam'] if c in df_train.columns]
    
    try:
        preprocessor_reg = ColumnTransformer(
            transformers=[
                ('num', StandardScaler(), num_features),
                ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_features)
            ],
            remainder='drop'
        )
    except TypeError:
        preprocessor_reg = ColumnTransformer(
            transformers=[
                ('num', StandardScaler(), num_features),
                ('cat', OneHotEncoder(handle_unknown='ignore', sparse=False), cat_features)
            ],
            remainder='drop'
        )
    
    # Drop rows with missing target
    train_reg = df_train.dropna(subset=[target]).copy()
    test_reg = df_test.dropna(subset=[target]).copy()
    print(f"Train rows with {target}: {train_reg.shape[0]}")
    print(f"Test rows with {target}: {test_reg.shape[0]}\n")
    
    if train_reg.shape[0] == 0:
        print("No training data; skipping regression")
    else:
        X_train_reg = train_reg[num_features + cat_features].copy()
        y_train_reg = train_reg[target]
        X_test_reg = test_reg[num_features + cat_features].copy()
        y_test_reg = test_reg[target]
        
        # Random Forest Regressor
        reg_rf = Pipeline([
            ('preprocessor', preprocessor_reg),
            ('model', RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1, max_depth=10))
        ])
        reg_rf.fit(X_train_reg, y_train_reg)
        y_pred_rf_reg = reg_rf.predict(X_test_reg)
        
        mae_rf = mean_absolute_error(y_test_reg, y_pred_rf_reg)
        rmse_rf = np.sqrt(mean_squared_error(y_test_reg, y_pred_rf_reg))
        r2_rf = r2_score(y_test_reg, y_pred_rf_reg)
        
        print("--- RANDOM FOREST ---")
        print(f"MAE:  {mae_rf:.4f}")
        print(f"RMSE: {rmse_rf:.4f}")
        print(f"R²:   {r2_rf:.4f}")
        
        # Gradient Boosting (often stronger)
        reg_gb = Pipeline([
            ('preprocessor', preprocessor_reg),
            ('model', GradientBoostingRegressor(n_estimators=100, random_state=42, max_depth=5))
        ])
        reg_gb.fit(X_train_reg, y_train_reg)
        y_pred_gb_reg = reg_gb.predict(X_test_reg)
        
        mae_gb = mean_absolute_error(y_test_reg, y_pred_gb_reg)
        rmse_gb = np.sqrt(mean_squared_error(y_test_reg, y_pred_gb_reg))
        r2_gb = r2_score(y_test_reg, y_pred_gb_reg)
        
        print("\n--- GRADIENT BOOSTING ---")
        print(f"MAE:  {mae_gb:.4f}")
        print(f"RMSE: {rmse_gb:.4f}")
        print(f"R²:   {r2_gb:.4f}")
        
        # Feature importance
        print("\n--- FEATURE IMPORTANCE (GB) ---")
        fi_reg = pd.DataFrame({
            'feature': num_features + cat_features,
            'importance': reg_gb.named_steps['model'].feature_importances_[:len(num_features + cat_features)]
        }).sort_values('importance', ascending=False).head(15)
        display(fi_reg)
        
        # Residual plot
        print("\n--- RESIDUALS (GB) ---")
        residuals = y_test_reg - y_pred_gb_reg
        print(f"Mean residual: {residuals.mean():.4f}")
        print(f"Std residual:  {residuals.std():.4f}")
        print(f"Min residual:  {residuals.min():.4f}")
        print(f"Max residual:  {residuals.max():.4f}")
```
